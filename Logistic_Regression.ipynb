**Question 1 : What is Simple Linear Regression (SLR)? Explain its purpose.**

Ans: Simple Linear Regression (SLR) is a statistical method used to study the relationship between two variables ‚Äî one independent variable (X) and one dependent variable (Y).

Purpose of SLR:

Prediction: To predict the value of one variable (Y) based on the known value of another (X).

Understanding Relationships: To determine whether and how strongly two variables are related.

Trend Analysis: To identify and represent linear trends in data.

Quantification: To measure the rate of change ‚Äî how much Y changes when X changes by one unit.

**Question 2: What are the key assumptions of Simple Linear Regression?**

Ans: Key Assumptions of Simple Linear Regression (SLR):

To ensure the model‚Äôs predictions and interpretations are valid, SLR relies on several key statistical assumptions:

1. Linearity: There must be a linear relationship between the independent variable (X) and the dependent variable (Y).



*   This means that changes in X are associated with proportional changes in Y.
*   A scatter plot of X vs. Y should roughly form a straight-line pattern.

2. Independence of Errors: The residuals (errors) ‚Äî the differences between observed and predicted values ‚Äî must be independent of each other.



*   In other words, one observation‚Äôs error should not influence another‚Äôs.
*   This is especially important in time series or sequential data.

3. Homoscedasticity (Constant Variance): The variance of the residuals should remain constant across all levels of X.




*   If the spread of residuals increases or decreases with X, it indicates heteroscedasticity, which can make the model unreliable.

4. Normality of Errors: The residuals should be normally distributed (bell-shaped curve) around zero.



*   This is important for hypothesis testing and constructing confidence intervals.

5. No Multicollinearity (Automatically satisfied in SLR): Since SLR has only one independent variable, this assumption is inherently met.



*   In multiple regression, this refers to independent variables not being highly correlated.

6. No Significant Outliers: There should be no extreme data points that disproportionately influence the regression line.



*   Outliers can distort the slope and intercept, leading to misleading conclusions.

**Question 7: Name and briefly describe three common evaluation metrics for regression models**

Ans: Three Common Evaluation Metrics for Regression Models

To evaluate how well a regression model fits and predicts data, several performance metrics are used. The three most common ones are:

1. Mean Absolute Error (MAE)

Definition:
MAE measures the average absolute difference between the actual values and the predicted values.

  image.png

  Interpretation:

It shows how much the predictions deviate from the actual values on average.

Lower MAE indicates better model performance.

It treats all errors equally, regardless of direction (positive or negative).

Example: If MAE = 5, it means predictions are off by 5 units on average.

2. Mean Squared Error (MSE)

Definition:
MSE measures the average of the squared differences between the actual and predicted values.

image.png

Interpretation:

Penalizes larger errors more than smaller ones (because of squaring).

Lower MSE means the model fits the data better.

Useful for comparing models ‚Äî but sensitive to outliers.

Example: If MSE = 16, the average squared error between predictions and true values is 16.

3. R-squared (Coefficient of Determination)

Definition:
R¬≤ measures the proportion of variance in the dependent variable (Y) that can be explained by the independent variable(s) in the model.

image.png

Interpretation:

R¬≤ ranges from 0 to 1.

Higher R¬≤ indicates a better fit (more variance explained by the model).

R¬≤ = 0 ‚Üí Model explains none of the variation; R¬≤ = 1 ‚Üí Model explains all variation.

Example: If R¬≤ = 0.85, the model explains 85% of the variation in the dependent variable.

**Question 8: What is the purpose of the R-squared metric in regression analysis?**

Ans: Purpose of the R-squared Metric in Regression Analysis

R-squared (R¬≤), also known as the coefficient of determination, is a key statistical measure used to evaluate how well a regression model fits the observed data.

Definition:

R¬≤ represents the proportion of the variance in the dependent variable (Y) that is explained by the independent variable(s) (X) in the model.

image.png

  Purpose and Interpretation:

  1. Measures Goodness of Fit:

Indicates how well the regression line fits the data.

A higher R¬≤ means the model explains a greater proportion of the variation in Y.

  2. Explains Variance:

Tells us what percentage of the dependent variable‚Äôs variation is explained by the independent variable(s).

Example: R¬≤ = 0.80 ‚Üí 80% of the variation in Y is explained by X; 20% remains unexplained.

  3. Model Comparison:

Helps compare multiple regression models ‚Äî the one with a higher R¬≤ generally fits the data better (assuming other conditions are constant).

  4. Indicates Predictive Power (with caution):

A high R¬≤ suggests better prediction ability, but it doesn‚Äôt guarantee that the model is correct or unbiased.

**Question 9: Write Python code to fit a simple linear regression model using scikit-learn and print the slope and intercept. (Include your Python code and output in the code box below.)**

Ans:  

    # Importing necessary libraries
    from sklearn.linear_model import LinearRegression
    import numpy as np

    # Sample data
    # Independent variable (X) - reshaped to 2D array as required by scikit-learn
    X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
    # Dependent variable (Y)
    Y = np.array([2, 4, 5, 4, 5])

    # Create a Linear Regression model
    model = LinearRegression()

    # Fit the model to the data
    model.fit(X, Y)

    # Get the slope (coefficient) and intercept
    slope = model.coef_[0]
    intercept = model.intercept_

    # Print results
    print("Slope (Œ≤‚ÇÅ):", slope)
    print("Intercept (Œ≤‚ÇÄ):", intercept)

Output:

    Slope (Œ≤‚ÇÅ): 0.6
    Intercept (Œ≤‚ÇÄ): 2.2

Explanation:

The LinearRegression() model from scikit-learn is used to fit a simple linear regression.

coef_ gives the slope (Œ≤‚ÇÅ) of the regression line.

intercept_ gives the intercept (Œ≤‚ÇÄ), i.e., the predicted value of Y when X = 0.
